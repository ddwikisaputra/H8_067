```python
data = pd.read_csv("property_data.csv")
data
```

```python
data['ST_NUM'].isnull() #cari data kosong
```

```python
newValues = []
for kData, vData in enumerate(data['OWN_OCCUPIED']):
    try :
        int(vData)
        newValues.append(np.nan)
    except ValueError :
         newValues.append(vData)

data['OWN_OCCUPIED'] = newValues
data.head(9)
```

```python
print('KOLOM : ', data.isnull().values.any()) #cek semua kolom yg ada value nan
print('JUMLAH : ', data.isnull().sum().sum()) #jumlah semua kolom yg ada value nan
```

```python
from datetime import datetime
```

```python
dateRange = pd.date_range(start = '1/01/2020', end = '1/8/2020', freq='H')
dateRange
```

```python
df = pd.DataFrame(columns=['date'], data=dateRange)
display(df)

df['data'] = np.random.randint(low=0,high=100, size=(len(dateRange)))
display(df)

df['datetime'] = pd.to_datetime(df['date'])
display(df)

df = df.set_index('datetime')
display(df)

df.drop(columns=['date'], inplace=True)
display(df)
```

```python
dfDate2 = df[df.index.day == 2]
display(dfDate2)
```

```python
dataPerDay = df.resample('D').mean() #pengelompokkan data per day
display(dataPerDay)
```

```python
df['rolling_sum'] = df.rolling(3).sum()
display(df)
```

```python
dateString = [str(x) for x in dateRange] #datetime to string

dateString
```

```python
opsdDaily = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/opsd_germany_daily.csv', index_col=0, parse_dates=True)
display(opsdDaily)

opsdDaily['Year'] = opsdDaily.index.year
opsdDaily['Month'] = opsdDaily.index.month
opsdDaily['Weekday'] = opsdDaily.index.weekday
display(opsdDaily)
```

```python
%matplotlib inline
opsdDaily['Consumption'].plot(title='Konsumsi Listrik Per Tahun', linewidth=0.5)
```

```python
%matplotlib inline
opsdDaily['Consumption'].plot(title='Konsumsi Listrik, Solar & Angin Per Tahun', marker='.', linestyle='None', alpha=0.5, figsize=(11,9))
opsdDaily['Solar'].plot(marker='.', linestyle='None', alpha=0.5, figsize=(11,9))
opsdDaily['Wind'].plot(marker='.', linestyle='None', alpha=0.5, figsize=(11,9))
```

```python
%matplotlib inline

print("TAHUN 2017")
ax = opsdDaily.loc['2017', 'Consumption'].plot(figsize=(11,9))
ax.set_ylabel('Daily Consumption (Gwh)')
```

```python
print("TAHUN 2017-01 s/d 2017-02")
ax = opsdDaily.loc['2017-01' : '2017-02', 'Consumption'].plot(marker='o', figsize=(11,9))
ax.set_ylabel('Daily Consumption (Gwh)')
```

```python
opsdDaily.boxplot(column=['Consumption', 'Solar', 'Wind'], by='Month')
```
